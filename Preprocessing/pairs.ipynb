```python
import pandas as pd
import matplotlib.pyplot as plt
from libs.pairs import pairs
```

```python
df = pd.read_csv('/Users/jonniebarnsley/code/phd/Preprocessing/PPE_design.csv')
parameter_names = ['gamma0', 'UMV', 'LRP', 'PDDi', 'WeertC']
parameters = df[parameter_names]
```

```python
ticks = {
    'gamma0'    : [0, 2e5, 4e5],
    'UMV'       : [0, 1e21],
    'LRP'       : [0, 4e-4, 8e-4],
    'PDDi'      : [1e-2, 2e-2],
    'WeertC'    : [1e4, 3e4, 5e4]
}

labels = ['$\gamma_0$', 'UMV', 'LRP', 'PDDi', 'C']

scatter_kw = {
    'edgecolor' : 'black',
    'facecolor' : 'none',
    's'         : 5,
    'linewidth' : 0.5
}

fig = pairs(parameters, ticks=ticks, labels=labels, **scatter_kw)
plt.savefig('./plots/pairsplot.svg', format='svg')
```
